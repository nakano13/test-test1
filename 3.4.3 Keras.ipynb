```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD

model = Sequential([
    Dense(input_dim=2, units=1),
    Activation("sigmoid")
])

# Sequential()は、層構造のモデルを定義するメソッド。ここに実際の層を入れ込むことで、モデルの設定をする。
# Dense(input_dim=2, units=1)により、入力の次元2、出力の次元1のネットワーク構造を持つ層を生成するが、これは数式における
# w₁x₁ + w₂x₂ + b

# ニューロンの出力を表現するには、ここの活性化関数が必要になるため、Activation("sigmoid")により、

# y = σ(w₁x₁ + w₂x₂ + b)
# を表現する層を作る。
# Sequential()だけを事前に宣言しておき、後から model.add()でどんどん層を追加していく書き方も可能。
# その場合、下記のようになる。

model = Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation("sigmoid"))
```

```python
# 確率的勾配降下法は下記の1行のみで表すことが可能。
model.compile(loss="binary_crossentropy", optimizer=SGD(lr=0.1))

# lrは学習率(learning rate)を表す。
```

```python
# モデルの学習も、Kerasでは1行書くのみ。ORゲートの入力・出力の正解データを用意し、

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])
```

```python
# model.fit ()を実行するだけ。
model.fit(X, Y, epochs=200, batch_size=1)

# epochsはエポック数、batch_sizeは(ミニ)バッチの大きさをそれぞれ指定。
# これだけでロジスティック回帰の学習まで実装できる。
```

```python
# 学習結果を確認する方法
classes = model.predict_classes(X, batch_size=1)
prob = model.predict_proba(X, batch_size=1)

print(classes)

print(prob)

# すると、分類されたパターン(=ニューロンが発火するかしないか)および、出力の確立が得られる。
# Kerasでは、シンプルに実装をまとめることが可能で、とりあえずモデルを試したい、
# 簡易的な実験を行いたいという場合にすばやく実装可能。
```

```python
# 全体のコードは以下の通り
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD

np.random.seed(0) # 乱数シード

# ・・・
# モデル設定
# ・・・

model = Sequential([
    Dense(input_dim=2, units=1),
    Activation("sigmoid")
])

# model = Sequential()
# model.add(Dense(input_dim=2, units=1))
# model.add(Activation("sigmoid"))

model.compile(loss="binary_crossentropy", optimizer=SGD(lr=0.1))

# ・・・
# モデル学習
# ・・・

# ORゲート
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])

model.fit(X, Y, epochs=200, batch_size=1)

# ・・・
# 学習結果確認
# ・・・

classes = model.predict_classes(X, batch_size=1)
prob = model.predict_proba(X, batch_size=1)

print("classified:")
print(Y == classes)
print()
print("output probability:")
print(prob)
```
